```python
X = [1,2,3,4,5]
Y = [10,11,12,13,14]

plt.plot(X, Y)
plt.show()
```

```python
X = [1,2,3,4,5]
Y = [10,11,12,13,14]

plt.plot(X, Y, marker = 'o')
plt.show()
```

```python
X = [1,2,3,4,5]
Y = [10,11,12,13,14]

plt.plot(X, Y, marker = 'o', color = 'green', linestyle = 'dotted')
plt.title("My Plot!")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid()
plt.show()
```
